```python
# Kredi Başvurusu Değerlendirme
print(">>>  Kredi Başvurusu Değerlendirme <<<")

try:
    yas = int(input("Yaşınızı girin: "))
    gelir = float(input("Aylık gelirinizi (TL) girin: "))

    if yas < 18:
        # Yaş 18'den küçükse
        print(f"Yaşınız {yas} olduğundan: **'Yaşınız kredi için uygun değil.'**")

    elif 18 <= yas <= 65:
        # Yaş 18 ile 65 arasındaysa
        if gelir > 5000:
            print(f"Yaşınız {yas}, Geliriniz {gelir:.2f} olduğundan: **'Kredi onaylandı.'**")
        else:
            print(f"Yaşınız {yas}, Geliriniz {gelir:.2f} olduğundan: **'Geliriniz yetersiz.'**")

    elif yas > 65:
        # Yaş 65'ten büyükse
        print(f"Yaşınız {yas} olduğundan: **'Yaşınız kredi için uygun değil.'**")

    else:
        
        print("Geçerli bir yaş aralığı girilmedi.")

except ValueError:
    print("**HATA: Lütfen geçerli sayısal değerler (Yaş için tam sayı, Gelir için sayı) girin.**")

print("-\n")
```
