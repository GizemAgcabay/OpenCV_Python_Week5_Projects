```python
#  Sıcaklık Analizi
print(">>> Sıcaklık Analizi<<<")

# Verilen sıcaklık listesi 
sicakliklar = [25.5, 25.8, 21.8, 19.2, 20.2, 22.7, 24.3, 
               25.6, 26.2, 25.8, 24.9, 23.7, 22.4, 21.1, 
               20.8, 19.9, 18.7, 19.2, 20.5, 21.9, 23.3, 
               22.8, 21.5]

# 1. Günlük ortalama sıcaklık hesaplama

toplam_sicaklik = sum(sicakliklar)
gun_sayisi = len(sicakliklar)
ortalama_sicaklik = toplam_sicaklik / gun_sayisi

print(f"Toplam Gün Sayısı: {gun_sayisi}")
print(f"Günlük Ortalama Sıcaklık: {ortalama_sicaklik:.2f} derece")

# 2. Günün en yüksek ve en düşük sıcaklığını bulma
en_yuksek = max(sicakliklar)
en_dusuk = min(sicakliklar)

print(f"Günün En Yüksek Sıcaklığı: {en_yuksek} derece")
print(f"Günün En Düşük Sıcaklığı: {en_dusuk} derece")

# 3. Konfor Analizi 
konfor_baslangic = 20
konfor_bitis = 24
konforlu_gunler = 0

for sicaklik in sicakliklar:
    if konfor_baslangic <= sicaklik <= konfor_bitis:
        konforlu_gunler += 1

print(f"\nKonfor Aralığı (20-24 derece) Analizi:")
print(f"Konforlu Gün Sayısı: {konforlu_gunler} gün")

# 4. Enerji Tüketim Uyarısı Analizi 
enerji_uyari_esigi = 18
uyari_gunleri = 0

for sicaklik in sicakliklar:
    if sicaklik < enerji_uyari_esigi:
        uyari_gunleri += 1

print(f"Enerji Tüketim Uyarısı (18 derece altı):")
print(f"Kimyasal Isıtmanın Gerekli Olduğu Gün Sayısı: {uyari_gunleri} gün")

print("\n")
```
