```python
# range() Fonksiyonu Kullanımı
print("---  range() Fonksiyonu Uygulamaları ---")

# a) 1'den 100'e kadar olan çift sayıların toplamı
print("a) 1'den 100'e Çift Sayıların Toplamı:")


cift_sayilarin_toplami = 0
for sayi in range(2, 101, 2):
    cift_sayilarin_toplami += sayi

print(f"Toplam: {cift_sayilarin_toplami}")
print("\n")



print("b) 50'den Geriye Doğru 40'a Kadar Sayılar:")


geri_sayim_listesi = []
for sayi in range(50, 39, -1):
    geri_sayim_listesi.append(sayi)


print(", ".join(map(str, geri_sayim_listesi))) 

print("\n")
```
